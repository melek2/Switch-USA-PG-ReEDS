```python
import pandas as pd
import re
import yaml
from collections import Counter

# --- edit these ---
xlsx_path = "pg/hierarchy_withanalysis.xlsx"
sheet_name = "Sheet2"
agg_col = "region_name"
ba_col  = "BAs (comma sep)"
out_path = "pg/region_definitions.yml"
# ------------------

df = pd.read_excel(xlsx_path, sheet_name=sheet_name)
df = df.dropna(subset=[agg_col, ba_col])

# Check for duplicate region names
names = [str(n).strip() for n in df[agg_col]]
dupes = [n for n, c in Counter(names).items() if c > 1]
if dupes:
    raise ValueError(f"Duplicate region_name entries found: {dupes}")

def natkey(s):
    return [int(t) if t.isdigit() else t for t in re.split(r'(\d+)', str(s))]

agg_to_bas = {}
for _, row in df.iterrows():
    name = str(row[agg_col]).strip()
    bas = [b.strip() for b in str(row[ba_col]).split(",") if b.strip()]
    agg_to_bas[name] = sorted(bas, key=natkey)

data = {
    "model_regions": sorted(agg_to_bas.keys(), key=natkey),
    "region_aggregations": agg_to_bas,
}

with open(out_path, "w") as f:
    yaml.safe_dump(data, f, default_flow_style=False, sort_keys=False)

print(f"Wrote {out_path}")
```
